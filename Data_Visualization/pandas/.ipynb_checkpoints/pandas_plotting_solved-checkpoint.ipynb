```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
raw_data = load_iris()
df = pd.DataFrame(data=raw_data.data, columns=raw_data.feature_names)
df["target"] = raw_data.target
df.columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", "target"]
```

```python
df.head(2)
```

```python
df.shape
```

# 1. Histogram

`df[feature_list].plot(kind="hist", bins=XX, alpha=X.X)`

Example: use feature list of sepal_length and petal_length

```python
# Code Block
df[["sepal_length","petal_length"]].plot(kind="hist", bins=50, alpha=0.5, figsize=(15, 5))
plt.title("Histogram of Lengths")
plt.xlabel("Mesurement")
plt.ylabel("Count")
plt.show()
```

 # 2. Density Curve
 
 `df[feature_list].plot(kind="kde", alpha=X.X)`
 
 Example: use feature list of sepal_length and petal_length

```python
# Code Block
df[["sepal_length","petal_length"]].plot(kind="kde", alpha=1, figsize=(15, 5))
plt.title("KDE curve")
plt.xlabel("Measurement")
plt.ylabel("Count")
plt.legend(loc=5)
plt.show()
```

# 3. Bar Chart
`df.plot(kind="bar", stacked=True)`

```python
avg_df = df.groupby(by='target', as_index=False).mean()
bar_df = avg_df.drop('target',axis=1)
bar_df.head()
```

```python
# Code Block
bar_df.plot(kind="bar", figsize=(10, 5))    # stacked=True
plt.title("Bar Chart")
plt.xlabel("Targets")
plt.ylabel("Measurements")
plt.show()
```

# 4. Pie Chart
`df[column_name].plot(kind="pie")`

```python
pie_df = df.target.value_counts()
pie_df
```

```python
# Code Block
pie_df.plot(kind="pie", figsize=(5, 5))
plt.title("Count of Targets")
plt.show()
```

# 5. Scatter Plot
`df.plot(kind="scatter",x="numerical_feature_1",y="numerical_feature_2", c="nominal_attribute", colormap="colormap_type")`
Example: use num_cols="sepal_length", "petal_length"; use color c="target" and colormap="Set1" 

```python
# Code Block
df.plot(kind="scatter", x="sepal_length", y="petal_length", c="target", colormap="Set1", figsize=(15, 5))
plt.title("Scatterplot")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
```

# 6. Box Plot
`df.plot(kind="box")`

```python
box_df=df.drop("target", axis=1)
box_df.head(2)
```

```python
# Code Block
box_df.plot(kind="box", figsize=(15, 6))
plt.title("Boxplot of Numeric Columns")
plt.xlabel("Name of Cols")
plt.ylabel("Measuement of Items")
plt.show()
```

# 7. Area Plot
`area_df.plot(kind="area")`

```python
area_df=df.drop("target", axis=1)
area_df.head(2)
```

```python
# Code Block
area_df.plot(kind="area", figsize=(15, 5))
plt.title("Area Plot")
plt.xlabel("Observations 0 to 150")
plt.ylabel("Measurement")
plt.show()
```

# 8. Line Plot
`df[col_name].plot(kind="line")`

Example: plot against any numeric column from "sepal_length", "sepal_width", "petal_length", "petal_width"

```python
# Code Block
df.sepal_length.plot(kind="line", figsize=(15, 5))
df.sepal_width.plot(kind="line", figsize=(15, 5))
df.petal_length.plot(kind="line", figsize=(15, 5))
plt.title("Lineplot")
plt.xlabel("Observation 0 to 150")
plt.ylabel("Measurements")
plt.legend()
plt.show()
```
