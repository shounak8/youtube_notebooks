```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
raw_data = load_iris()
df = pd.DataFrame(data=raw_data.data, columns=raw_data.feature_names)
df['target'] = raw_data.target
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'target']
```

```python
df.head()
```

# 1. Histogram

```python
df[["sepal_length","petal_length"]].plot(kind="hist", bins=20, alpha=0.6)
```

 # 2. Density Curve
 
 df[feature_list].plot(kind="kde", alpha=X.X)

```python
df[["sepal_length","petal_length"]].plot(kind="kde")
```

# 3. Bar Chart
df.plot(kind="bar")

```python
avg_df = df.groupby(by='target', as_index=False).mean()
avg_df
```

```python
avg_df.drop('target',axis=1).plot(kind="bar", stacked=True)
```

# 4. Pie Chart
df[column_name].plot(kind="pie")

```python
df.target.value_counts().plot(kind="pie")
```

# 5. Scatter Plot
df.plot(kind="scatter",x="numerical_feature_1",y="numerical_feature_2", c="nominal_attribute", colormap="colormap_type")

```python
df.plot(kind="scatter",x="sepal_length",y="sepal_width", c="target", colormap="Set1")
```

# 6. Box Plot
df.plot(kind="box")

```python
df.drop("target", axis=1).plot(kind="box")
```

# 7. Area Plot

```python
df.drop("target", axis=1).plot(kind="area")
```

# 8. Line Plot

```python
df.sepal_length.plot(kind="line")
```
